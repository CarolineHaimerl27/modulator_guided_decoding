```python
PERF = np.load('Fig4E_PERF.npy')
NTRAINreal = np.load('Fig4E_TRAIN.npy')
```

```python
decoder = ['opt dec', 'MG dec', 'regression']

######### parametrized ############

MEAN_chance = np.zeros([PERF.shape[0], len(decoder), 2]) * np.nan
fig, ax = plt.subplots(1,1, figsize=(4,4))

for dd in range(len(decoder)):
    perf = PERF[:, :, dd, :]
    ntrain = NTRAINreal[:, dd, :]
    for bb in range(perf.shape[0]):  # all blocks of that monkey

        # when first above chance
        FIRST = np.zeros(perf.shape[1]) * np.nan
        # look in each fold where it first surpassed 50%
        for ss in range(perf.shape[1]):  # each fold
            tmp = np.where(perf[bb, ss, :] > 50)[0]
            if len(tmp) > 0:
                FIRST[ss] = tmp[0]
        # set those to last training
        notreached = np.isnan(FIRST)
        FIRST[notreached] = 17
        ntrain_bb = ntrain[bb, FIRST.astype('int')]
        ntrain_bb[notreached] = ntrain_bb[notreached] + 10  # ARBITRARY!!!
        perf_bb = perf[bb, :, :]
        perf_bb_tmp = np.zeros([perf_bb.shape[0], perf_bb.shape[1]]) * np.nan
        for tt in range(perf_bb_tmp.shape[0]):
            perf_bb_tmp[tt,:(FIRST[tt].astype('int')+1)] = perf_bb[tt, :(FIRST[tt].astype('int')+1)]
        ax.plot(np.mean(ntrain_bb), np.mean(np.nanmean(perf_bb_tmp, axis=1)), '.', color=cols_decode[decoder[dd]])

        MEAN_chance[bb, dd, 0] = np.mean(ntrain_bb)
        MEAN_chance[bb, dd, 1] = np.mean(np.nanmean(perf_bb_tmp, axis=1))

for dd in range(len(decoder)):

    ax.plot(np.nanmean(MEAN_chance[:, dd, 0]), np.nanmean(MEAN_chance[:, dd, 1]), 'o',
                  markersize=10, color=cols_decode[decoder[dd]])


    ax.set_xscale('log')
ax.set_xticks([3, 10, 40])
ax.set_xticklabels([3, 10, 40])
ax.legend()
ax.set_xlim(3.5,10)

fig.tight_layout()

ax.set_ylim(48,80)
ax.set_xlabel('#trials to reach criterion')
ax.set_ylabel('performance')
```
