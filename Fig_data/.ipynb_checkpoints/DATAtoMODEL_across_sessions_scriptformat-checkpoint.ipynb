```python
def my_hist2(h1, h2, label1, label2, ax=None,
             color='k', legend=True):
    bins = h1[1]
    width = np.diff(bins)-np.diff(bins)*.3
    if ax is None:
        fig, ax, _ = myplot(1)
    ax.fill_between(np.array([h1[1][:-1], h1[1][1:]]).T.ravel(),
                    np.zeros(2*len(h1[0])),
                    np.repeat(h1[0]/np.sum(h1[0]),2), color=color, label=label1, alpha=.4)

    ax.plot(np.array([h2[1][:-1], h2[1][1:]]).T.ravel(),
                np.repeat(h2[0]/np.sum(h2[0]),2), linewidth = 2,
               color=color, label=label2)
    if legend: ax.legend()
```

### task3: overall modulation strength

is the strength of the modulator different in task 1/2 (relevant stimulus is in receptive field) and task 3 (relevant stimulus outside of receptive field).

For this I take the product of the norm of the modulation targeting parameter in the model (C) and the norm of the latent variable (x).

```python
df = pd.read_excel(r'Fig_4AB.xlsx')

fig, ax = plt.subplots(1,1, figsize=(4,4))
bins = np.logspace(-5, 2, 20, base=10)

h1 = np.histogram(df.loc[((df.task == 2) | (df.task == 1)), 'prod'], bins=bins)
h2 = np.histogram(df.loc[(df.task == 3), 'prod'], bins=bins)
my_hist2(h1, h2, label1='task1&2', label2='task3',
         color='purple', ax=ax)

ax.plot([np.nanmean(df.loc[((df.task == 2) | (df.task == 1)), 'prod']),
            np.nanmean(df.loc[((df.task == 2) | (df.task == 1)), 'prod'])],
           [0, np.max(h1[0] / np.sum(h1[0]))], '--', color='k', linewidth=3)
ax.plot([np.nanmean(df.loc[(df.task == 3), 'prod']),
            np.nanmean(df.loc[(df.task == 3), 'prod'])],
           [0, np.max(h1[0] / np.sum(h1[0]))], '-', color='k', linewidth=3)

ax.set_xscale('log')
ax.set_xticks(np.array([.1, 10, 10 ** 4]));
ax.legend()
ax.set_xlabel('var from mod versus stim')  # |x|*|C|')


############ same for task 1/2 comparison ##############
fig, ax = plt.subplots(1,1, figsize=(4,4))
bins = np.logspace(-5, 2, 20, base=10)

h1 = np.histogram(df.loc[(df.task == 1), 'prod'], bins=bins)
h2 = np.histogram(df.loc[(df.task == 2), 'prod'], bins=bins)
my_hist2(h1, h2, label1='task1', label2='task2',
         color='purple', ax=ax)

ax.plot([np.mean(df.loc[(df.task == 1), 'prod']),
            np.mean(df.loc[(df.task == 1), 'prod'])],
           [0, np.max(h1[0] / np.sum(h1[0]))], '--', color='k', linewidth=3)
ax.plot([np.mean(df.loc[(df.task == 2), 'prod']), np.mean(df.loc[(df.task == 2), 'prod'])],
           [0, np.max(h1[0] / np.sum(h1[0]))], '-', color='k', linewidth=3)

ax.set_xscale('log')
ax.set_xticks(np.array([.1, 10, 10 ** 4]));
ax.legend()
ax.set_xlabel('var from mod versus stim') 
```

```python
cols_decode = {'opt dec': 'red', \
           'MG dec': 'green',\
           'RG dec': 'blue', 'SO dec': 'grey', 'sum dec': 'orange', 'regression': 'orange', 'svm': 'purple'}
```

```python
df = pd.read_excel(r'Fig_4D.xlsx')

fig, ax = plt.subplots(1,1, figsize=(4,4))
bins = np.linspace(0, 100, 20)
decoder = ['opt dec', 'regression', 'MG dec']
for dd in decoder:  # decoder
    print('decoder: ', dd)
    meandd = df[dd+'_mean']  #np.array([np.nanmean(perf[nn]) for nn in range(len(perf))])
    vardd = df[dd+'_sd'] 
    ax.plot(df['train'], meandd, '-', color=cols_decode[dd], label=dd)

ax.plot([np.nanmin(df['train']), np.nanmax(df['train'])], [50, 50], '--k')
ax.set_ylim(0, 100)
ax.set_xlabel('# training')
#ax.set_xscale('log')
ax.set_xlim(0,70)
ax.set_xticks([np.nanmin(df['train']), 10, 20, 40, 60])
ax.set_xticklabels([np.nanmin(df['train']), 10, 20, 40, 60])
ax.set_ylim(40, 100)
ax.set_ylabel('Performance both')
fig.tight_layout()
```

```python
PERF = np.load('Fig4E_PERF.npy')
NTRAINreal = np.load('Fig4E_TRAIN.npy')
decoder = ['opt dec', 'MG dec', 'regression']

######### parametrized ############

MEAN_chance = np.zeros([PERF.shape[0], len(decoder), 2]) * np.nan
fig, ax = plt.subplots(1,1, figsize=(4,4))

for dd in range(len(decoder)):
    perf = PERF[:, :, dd, :]
    ntrain = NTRAINreal[:, dd, :]
    for bb in range(perf.shape[0]):  # all blocks of that monkey

        # when first above chance
        FIRST = np.zeros(perf.shape[1]) * np.nan
        # look in each fold where it first surpassed 50%
        for ss in range(perf.shape[1]):  # each fold
            tmp = np.where(perf[bb, ss, :] > 50)[0]
            if len(tmp) > 0:
                FIRST[ss] = tmp[0]
        # set those to last training
        notreached = np.isnan(FIRST)
        FIRST[notreached] = 17
        ntrain_bb = ntrain[bb, FIRST.astype('int')]
        ntrain_bb[notreached] = ntrain_bb[notreached] + 10  # ARBITRARY!!!
        perf_bb = perf[bb, :, :]
        perf_bb_tmp = np.zeros([perf_bb.shape[0], perf_bb.shape[1]]) * np.nan
        for tt in range(perf_bb_tmp.shape[0]):
            perf_bb_tmp[tt,:(FIRST[tt].astype('int')+1)] = perf_bb[tt, :(FIRST[tt].astype('int')+1)]
        ax.plot(np.mean(ntrain_bb), np.mean(np.nanmean(perf_bb_tmp, axis=1)), '.', color=cols_decode[decoder[dd]])

        MEAN_chance[bb, dd, 0] = np.mean(ntrain_bb)
        MEAN_chance[bb, dd, 1] = np.mean(np.nanmean(perf_bb_tmp, axis=1))

for dd in range(len(decoder)):

    ax.plot(np.nanmean(MEAN_chance[:, dd, 0]), np.nanmean(MEAN_chance[:, dd, 1]), 'o',
                  markersize=10, color=cols_decode[decoder[dd]])


    ax.set_xscale('log')
ax.set_xticks([3, 10, 40])
ax.set_xticklabels([3, 10, 40])
ax.legend()
ax.set_xlim(3.5,10)

fig.tight_layout()

ax.set_ylim(48,80)
ax.set_xlabel('#trials to reach criterion')
ax.set_ylabel('performance')
```
