```python
decnames = ['weights', 'gain', 'stoch', 'decoder']
dec_colors = {'stoch': 'green', 'weights': 'red', 'decoder': 'orange', 'gain': 'blue'}

fig2, ax2 = plt.subplots(1, 1, figsize=(4, 4))

for mm in [3,0, 2,1]:
    correct_mm = np.stack(correct)[:,mm,:(np.max(np.sum(np.isnan(np.stack(correct)[:,mm,:])==False,axis=1)))]*100
    if correct_mm.shape[1]>0:
        nsamp = np.sqrt(np.sum(np.isnan(correct_mm[:,0])==False))
        tmp_len = len(np.nanmean(correct_mm, axis=0))
        ax2.fill_between(np.arange(0, Tpar_tun['Nsamples'] * (tmp_len), Tpar_tun['Nsamples']),
                       np.nanmean(correct_mm, axis=0) - np.sqrt(np.nanvar(correct_mm, axis=0))/nsamp,
                       np.nanmean(correct_mm, axis=0) + np.sqrt(np.nanvar(correct_mm, axis=0))/nsamp,
                        alpha=.2, color=dec_colors[decnames[mm]]);
for mm in [3,0, 2,1]:
    correct_mm = np.stack(correct)[:,mm,:(np.max(np.sum(np.isnan(np.stack(correct)[:,mm,:])==False,axis=1)))]*100
    if correct_mm.shape[1]>0:
        tmp_len = len(np.nanmean(correct_mm, axis=0))
        ax2.plot(np.arange(0, Tpar_tun['Nsamples'] * (tmp_len), Tpar_tun['Nsamples']),
               np.nanmean(correct_mm, axis=0), '-', color=dec_colors[decnames[mm]], linewidth=2, label=decnames[mm]);
ax2.plot([0, ax2.get_xlim()[1]], [50, 50], '--k')
ax2.legend()
ax2.set_title('% correct')
ax2.set_xlabel('# data points')
ax2.set_ylabel('accuracy (%)')
ax2.set_ylim(0,101) 
ax2.set_xlim(-10, 300)

fig2.tight_layout()
```
